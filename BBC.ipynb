```python
import requests
from bs4 import BeautifulSoup
```

```python
# Ensuite récupérer le contenu HTML de la page web BBC https://www.bbc.com
```

```python
page = requests.get("https://www.bbc.com")
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
# Puis , j'extraire le titre de la page
```

```python
title = soup.find('title').get_text()
```

```python
# Je l'affiche pour verifier
```

```python
title
```

```python
# J'initialiser une liste pour stocker les informations sur les articles
```

```python
articles_data = []
```

```python
# J'extraire tous les articles sur la page
```

```python
articles = soup.find_all('article')
```

```python
# Je les affiche pour vérifier
```

```python
articles
```

```python
# Je parcour chaque article pour extraire le titre de l'article, le lien de l'article , la description de l'article si il en a une 
```

```python
for article in articles:
    # Initialiser un dictionnaire pour stocker les informations sur un article
    article_info = {}
    
    # Extraire le titre de l'article
    article_title_element = article.find('h2', class_='sc-4fedabc7-3')
    if article_title_element:
        article_info['title'] = article_title_element.get_text()
    else:
        article_info['title'] = "Titre non disponible"
    
    # Extraire le lien de l'article
    article_link_element = article.find('a', class_='sc-4befc967-1')
    if article_link_element and 'href' in article_link_element.attrs:
        article_info['link'] = 'https://www.bbc.com' + article_link_element['href']
    else:
        article_info['link'] = "Lien non disponible"
    
    # Extraire la description de l'article
    article_description_element = article.find('p', class_='sc-1a9a6d5b-0')
    if article_description_element:
        article_info['description'] = article_description_element.get_text()
    else:
        article_info['description'] = "Description non disponible"
    
    # Ajouter les informations de l'article à la liste d'articles
    articles_data.append(article_info)
```

```python
# J'affiche les données que j'ai récupérées
```

```python
print("Titre de la page:", title)
print("Articles:")
for i, article in enumerate(articles_data, 1):
    print(f"Article {i}:")
    print("Titre:", article['title'])
    if 'link' in article:
        print("Lien:", article['link'])
    else:
        print("Lien non disponible")
    print("Description:", article['description'])
    print()
```
